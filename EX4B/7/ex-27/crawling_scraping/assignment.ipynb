### Sample program for crawling  

#### Import libraries  

```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#### Parameters  

```python
url = 'https://www.toyo.ac.jp/toyo2020/basic_data/'
csv_out = 'num_students_toyo.csv'
```

#### Crawling of web page  

```python
s = requests.Session()
r = s.get(url)

print(r.status_code)
print(r.encoding)
```

```python
# response body
print(r.text)  # str
#print(r.content)  # binary
```

#### Scraping (parse web contents)  

```python
bs = BeautifulSoup(r.text, 'html.parser')
```

#### Retrieve first table

```python
table_all = bs.select('.data-table')
table = table_all[0]
```

#### Filter headings (columns)   

```python
cols = []
for col in table.thead.find_all('th'):
    cols.append(col.text)
```

#### Filter data (rows)

```python
rows = []
for tr in table.tbody.find_all('tr'):
    row = []
    for th in tr.find_all('th'):
        row.append(th.text)
    for td in tr.find_all('td'):
        row.append(td.text)
    rows.append(row)
```

#### Output to CSV  

```python
df = pd.DataFrame(rows, columns=cols)
display(df)

df.to_csv(csv_out, index=False)
```
