```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def okuhata(f, hb, hm, d, city):
    if city == 'small': ahm = (1.1* math.log10(f)-0.7)*hm - (1.56*math.log10(f)-0.8)
    elif city =='large':
        if f < 400: ahm = 8.29*(math.log10(1.54*hm))**2 - 1.1
        elif f >= 400: ahm = 3.2*(math.log10(11.75*hm))**2 - 4.97
    return 69.55 + 26.16 * math.log10(f) - 13.82 * math.log10(hb) + (44.9 - 6.55*math.log10(hb))*math.log10(d) - ahm
```

```python
f = 800
hm = 1.5
city = 'small'
hb_arr = [30, 120, 180]
d_arr = [i for i in range(1,11)]
# d_arr = np.log10([i for i in range(1,11)])
```

```python
x = np.log10([i for i in range(1,11)])
for hb in hb_arr:
    loss = []
    for d in d_arr:
        loss.append(okuhata(f, hb, hm, d, city))
    print(loss)
    plt.plot(x, loss, label=hb)
    plt.legend()
```
